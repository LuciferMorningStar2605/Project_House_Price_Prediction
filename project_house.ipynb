```python
from sklearn import linear_model
```

```python
x1= houseprice[['bedrooms', 'bathrooms','sqft_living','sqft_lot','view','condition','floors','sqft_above','sqft_basement','yr_built','yr_renovated']]
y1= houseprice[['price']]

reg=linear_model.LinearRegression()
reg.fit(x1,y1)
print(reg)
```

```python
reg.score(x1,y1)
```

```python
y_p=reg.predict(x1)
```

```python
plt.scatter(houseprice['sqft_living'],y1, c='blue',label='Actual Data')
plt.plot(houseprice['sqft_living'],y_p,c='red')
```
